# Encryption Time Anaylsys

```python
import matplotlib.pyplot as plt

# Read data from file
with open('encryption_timing_analysis.txt', 'r') as file:
  data = file.readlines()
  data = [float(x.strip()) for x in data]

# Generate X-axis values
x_axis = [i for i in range(28, len(data) + 28)]

# Plot graph
plt.plot(x_axis, data, '-o')
plt.xlabel('Key Size (bits)')
plt.ylabel('Time (s)')
plt.title('RSA Encryption Time Analysis')
plt.show()
```
